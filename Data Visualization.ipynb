```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("a.csv").drop(["Timestamp","Name",],axis=1)
```

```python
data
```

```python
data.loc[(data.Age<=20),'Agegroup']='<=20'
data.loc[(data.Age<=30) & (data.Age>20),'Agegroup']='20-30'
data.loc[(data.Age<=40) & (data.Age>30),'Agegroup']='30-40'
data.loc[(data.Age<=50) & (data.Age>40),'Agegroup']='40-50'
data.loc[(data.Age>50),'Agegroup']='>50'
```

```python
data.drop(["Age"],axis=1)
```

```python
x=data.groupby('Agegroup')
l=[x.get_group(i) for i in x.groups]
```

```python
x=data.groupby(data['Agegroup'])['Agegroup'].count()
labels=list(set(data["Agegroup"]))
plt.pie(x,labels=labels)
```

```python
####     AGE WISE     #####

group_by_gender=data.groupby(['Agegroup','Gender'])
count_by_gender=group_by_gender.size().unstack()
count_by_gender
count_by_gender.plot(kind='pie',subplots=True)
```
